```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

disease_model = load_model("skin_disease_classifier.h5")         # Disease classifier
acne_severity_model = load_model("acne_severity_classifier.h5")  # Acne severity classifier

disease_labels = ['Vitiligo', 'acne', 'ringworm'] 
acne_severity_labels = ['mild', 'moderate', 'severe']

def predict_combined(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    
    disease_pred = disease_model.predict(img_array)
    disease_idx = np.argmax(disease_pred)
    disease_name = disease_labels[disease_idx]
    print(f"Disease prediction: {disease_name}")
    
    if disease_name == 'acne':
        severity_pred = acne_severity_model.predict(img_array)
        severity_idx = np.argmax(severity_pred)
        severity_name = acne_severity_labels[severity_idx]
        print(f"Acne severity prediction: {severity_name}")
        return disease_name, severity_name
    else:
        return disease_name, None

img_path = 'test_image17.jpg'
disease, severity = predict_combined(img_path)
print(f"Result: {disease} ({severity})")
```
